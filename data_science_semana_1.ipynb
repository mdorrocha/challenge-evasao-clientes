## Importação do arquivo de dados


```python
import pandas as pd
```

```python
pd.set_option("display.max_columns", 100)
```

```python
# leitura do arquivo json
dados = pd.read_json('dados/telco-customer-churn.json')
```

```python
dados.head()
```

```python
# números de linhas e colunas
print('Total de linhas: ' + str(dados.shape[0]))
print('Total de colunas: ' + str(dados.shape[1]))
```

## Normalização dos dados json

```python
# normalizando a coluna customer
customer = pd.json_normalize(data=dados.customer)
```

```python
# normalizando a coluna phone
phone = pd.json_normalize(data=dados.phone)
```

```python
# normalizando a coluna internet
internet = pd.json_normalize(data=dados.internet)
```

```python
# normalizando a coluna account
account = pd.json_normalize(data=dados.account)
```

```python
# juntando os dataframes
dados_combinados = pd.concat([dados, customer, phone, internet, account], axis=1)
```

```python
# eliminando colunas redundantes
dados_combinados = dados_combinados.drop(['customer', 'phone', 'internet', 'account'], axis=1)
dados_combinados.head()
```

## Análise do DataFrame

```python
# pesquisar dados nulos
dados_combinados.info()
```

```python
# analisando os dados das colunas
for column in dados_combinados.columns:
    print(column, dados_combinados[column].unique())
```

```python
# convertendo a coluna charges.total para float
dados_combinados['Charges.Total'] = pd.to_numeric(dados_combinados['Charges.Total'], errors='coerce').fillna(0.0)
```

```python
# total de registros cujo valor em Churn é vazio
churn_vazio = dados_combinados.Churn == ''
dados_combinados[churn_vazio].shape[0]
```

```python
# eliminando registros cujo valor em Churn é vazio
dados_combinados.drop(dados_combinados[churn_vazio].index, inplace=True)
dados_combinados
```

```python
# reindexando o dataframe
new_index = range(dados_combinados.shape[0])
dados_combinados.index = new_index
dados_combinados.info()
```

## Tradução das colunas

```python
#traduzindo as colunas
colunas_traduzidas = ['Id Cliente', 'Cancelado', 'Gênero', 'Maior igual a 65', 'Parceiro', 'Dependentes', 
'Meses de contrato', 'Serviço telefônico', 'Múltiplas linhas', 'Serviço de internet', 'Segurança online', 
'Backup online', 'Proteção do dispositivo', 'Suporte técnico', 'TV a cabo', 'Streaming de filmes', 
'Tipo de contrato', 'Fatura online', 'Forma de pagamento', 'Valor por mês', 'Valor total']
dados_combinados.columns = colunas_traduzidas
```

```python
# traduzindo os valores nas colunas
colunas_valores = {
    ('Cancelado', 'Parceiro', 'Dependentes', 'Serviço telefônico', 'Fatura online'): {
        'Yes': 'Sim',
        'No': 'Não'
    },
    ('Gênero', ): {
        'Female': 'Feminino',
        'Male': 'Masculino'
    },
    ('Serviço de internet', ): {
        'DSL': 'DSL',
        'Fiber optic': 'Fibra Ótica',
        'No': 'Não'
    },
    ('Segurança online', 'Backup online', 'Proteção do dispositivo', 
    'Suporte técnico', 'TV a cabo', 'Streaming de filmes'): {
        'No': 'Não',
        'Yes': 'Sim',
        'No internet service': 'Sem internet'
    },
    ('Tipo de contrato', ): {
        'One year': 'Anual',
        'Month-to-month': 'Mensal',
        'Two year': '2 anos'
    },
    ('Forma de pagamento', ): {
        'Mailed check': 'Cheque enviado',
        'Electronic check': 'Cheque eletrônico',
        'Credit card (automatic)': 'Cartão de crédito',
        'Bank transfer (automatic)': 'Transferência bancária'
    },
    ('Múltiplas linhas', ): {
        'No': 'Não',
        'Yes': 'Sim',
        'No phone service': 'Sem telefone'
    },
    ('Maior igual a 65', ): {
        0: 'Não',
        1: 'Sim'
    }
}
for colunas, value in colunas_valores.items():
    for coluna in colunas:
        dados_combinados[coluna] = dados_combinados[coluna].map(value)
dados_combinados.head()
```

## Inclusão de coluna extra

```python
# incluindo a coluna Gastos diários
dados_combinados.insert(19, 'Gastos diários', value=round(dados_combinados['Valor por mês']/30, 2))
dados_combinados
```

## Salvando o DataFrame

```python
# salvando o DataFrame
dados_combinados.to_csv('Dados/dados_tratados.csv', index=False)
```
